# Identifying unique values and NA

# First Childhood Dependency Data

```python
# Expected unique values columns
# Look for the columns where we expect few unique values and analyse them in contrast the info we expected of the ICBF dictionaries information
# print max values, min values, unique values, nan values
# first create separated variables for all the columns we could analyse
all_columns = tomas_DPI.columns
all_columns
```

```python
# Loop for going through all the columns and get important general information

for col in all_columns:
    try:
        # for numeralical features
        unique_values = tomas_DPI[col].unique().sort_values()
        nunique_values = tomas_DPI[col].nunique().sort_values()
        print(f" {col}: unique: {nunique_values} are {unique_values}")
    except:
        # For categorical features
        unique_values = tomas_DPI[col].unique()
        nunique_values = tomas_DPI[col].nunique()
        print(f" {col}: unique: {nunique_values} are {unique_values}")
    try:
        # For numerical features
        max_val = tomas_DPI[col].max()
        min_val = tomas_DPI[col].min()
        print(f"max val is {max_val} min val is {min_val}")
    except:
        print("")
    # nan count
    nan_count = tomas_DPI[col].isna().value_counts()
    print(f"NAN: {nan_count}")
    #value_counts
    value_count = tomas_DPI[col].value_counts()
    print(f"Value counts: \n {value_count}")
    print(f'\n\n')
```
