# Identifying unique values and NA

```python
# Expected unique values columns
# Look for the columns where we expect few unique values and analyse them in contrast the info we expected of the ICBF information
# print max values, min values, unique values, nan values
# first create separated variables for all the columns we could analyse
all_columns = soc.columns
filtered_columns = all_columns.drop(['ingresos_promP_imp','gasto_ppers_imp', 'gasto_alim_ppers_imp', 'porc_gasto_alim', 'union_temp', 'cod_mpio','cod_dpto', 'FECHANAC2', 'IdBeneficiario_tomas17-19','id'])
```

```python
filtered_columns
```

```python
rest_of_columns = ['ingresos_promP_imp','gasto_ppers_imp', 'gasto_alim_ppers_imp', 'porc_gasto_alim', 'union_temp', 'cod_mpio','cod_dpto', 'FECHANAC2', 'IdBeneficiario_tomas17-19','id']
```

```python
rest_of_columns
```

```python
for col in filtered_columns:
    try:
        # Numerical features
        unique_values = soc[col].unique().sort_values()
        nunique_values = soc[col].nunique().sort_values()
        print(f" {col}: unique: {nunique_values} are {unique_values}")
    except:
        # Categorical features
        unique_values = soc[col].unique()
        nunique_values = soc[col].nunique()
        print(f" {col}: unique: {nunique_values} are {unique_values}")
    try:
        # Numerical features
        max_val = soc[col].max()
        min_val = soc[col].min()
        print(f"max val is {max_val} min val is {min_val}")
    except:
        print("")
    nan_count = soc[col].isna().value_counts()
    print(f"NAN: {nan_count}")
    
    value_count = soc[col].value_counts()
    print(f"Value counts: \n {value_count}")
    print(f'\n\n')
```
